```python
import pandas as pd
from scipy.stats import kruskal, mannwhitneyu
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Projetos/Finch/Data/plot.csv', encoding='utf-8-sig')
```

```python
def diff_test(data, cat_var, quant_var):    
    sns.boxplot(x=cat_var, y=quant_var, data=df)
        
    ktest = round(kruskal(*[group[quant_var].values for name, group in data.groupby(cat_var)]).pvalue, 3)
    mw_bf = round( mannwhitneyu(*[group[quant_var].values for name, group in data.groupby(cat_var) if name == 'bossa_nova' or name == 'funk']).pvalue, 3)
    mw_bg = round( mannwhitneyu(*[group[quant_var].values for name, group in data.groupby(cat_var) if name == 'bossa_nova' or name == 'gospel']).pvalue, 3)
    mw_bs = round( mannwhitneyu(*[group[quant_var].values for name, group in data.groupby(cat_var) if name == 'bossa_nova' or name == 'sertanejo']).pvalue, 3)
    mw_fg = round( mannwhitneyu(*[group[quant_var].values for name, group in data.groupby(cat_var) if name == 'funk' or name == 'gospel']).pvalue, 3)
    mw_fs = round( mannwhitneyu(*[group[quant_var].values for name, group in data.groupby(cat_var) if name == 'funk' or name == 'sertanejo']).pvalue, 3)
    mw_gs = round( mannwhitneyu(*[group[quant_var].values for name, group in data.groupby(cat_var) if name == 'gospel' or name == 'sertanejo']).pvalue, 3)

    print(f"Comparando a variável {quant_var} entre todos os grupos obtivemos um p-valor de {ktest}")
    print(f"Comparando a variável {quant_var} entre a bossa nova e o funk obtivemos um p-valor de {mw_bf}")
    print(f"Comparando a variável {quant_var} entre a bossa nova e o gospel obtivemos um p-valor de {mw_bg}")
    print(f"Comparando a variável {quant_var} entre a bossa nova e o sertanejo obtivemos um p-valor de {mw_bs}")
    print(f"Comparando a variável {quant_var} entre a funk e o gospel obtivemos um p-valor de {mw_fg}")
    print(f"Comparando a variável {quant_var} entre a funk e o sertanejo obtivemos um p-valor de {mw_fs}")
    print(f"Comparando a variável {quant_var} entre a gospel e o sertanejo obtivemos um p-valor de {mw_gs}")
    
    return None
```

```python
diff_test(df, 'Class', 'word_count')
```

```python
diff_test(df, 'Class', 'sent_count')
```
